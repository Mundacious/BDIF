# 000. Data Plotting - Order Book Analysis

#### Import packages 

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import datetime
plt.style.use('seaborn-poster')
%matplotlib inline
```

#### CSV read and cleaning

```python
def CSV_read_clean(ticker):   
    #  Read and clean CSV of each file
    order_book_file_name = ticker + "_order.csv"
    message_book_file_name = ticker + "_message.csv"
    
    msg_book = pd.read_csv(message_book_file_name)
    msg_book.columns = ["Time", "Type", "OrderID", "Size", "Price", "TradeDirection"]
    
    order_book = pd.read_csv(order_book_file_name)
    column_names = []
    nb_levels = 10
    for j in range(1, nb_levels+1, 1):
        column_names.append("ASKp" + str(j))
        column_names.append("ASKs" + str(j))
        column_names.append("BIDp" + str(j))
        column_names.append("BIDs" + str(j))
    order_book.columns = column_names

    order_book[list(range(33,40))].head()
    
    trading_start_time = 9.5*60*60;       # 9:30:00.000 in ms after midnight
    trading_end_time = 16*60*60;        # 16:00:00.000 in ms after midnight

    msg_book = msg_book.loc[(msg_book['Time'] >= trading_start_time) & (msg_book['Time'] <= trading_end_time)]
    order_book = order_book.loc[(msg_book['Time'] >= trading_start_time) & (msg_book['Time'] <= trading_end_time)]
    
    price_columns = [i for i in order_book.columns.values if 'p' in i]
    order_book[price_columns] = order_book[price_columns].apply(lambda x: x/10000, axis=0)

    return order_book, msg_book
```

#### Plot evolution of Bid Ask prices over the day

```python
def plot_levels_evolution(order_book, msg_book,ax,ticker,nb_levels=3, start_index=1, end_index=1000,plot_mid_price=False):
    """
    Parameters
    ----------
    order_book
    msg_book
    nb_levels
    start_index
    end_index
    plot_mid_price

    Returns
    -------

    """
    if nb_levels < 0 or nb_levels > 6:
        print("Not enough colors to plot that many levels. Max 3")
        nb_levels = 2

    blues = sns.color_palette("Blues").as_hex()[::-1]
    reds = sns.color_palette("Reds").as_hex()[::-1]

    times = msg_book.loc[start_index:end_index, 'Time']

    for i in range(1, (nb_levels+1)):
        best_ask_prices = order_book.loc[start_index:end_index, 'ASKp' + str(i)]
        best_bid_prices = order_book.loc[start_index:end_index, 'BIDp' + str(i)]

        ax.step(times, best_ask_prices, color=reds[i],
                 linestyle='-', label='ASK' + str(i))
        ax.step(times, best_bid_prices, color=blues[i],
                 linestyle='-', label='BID' + str(i))

    if plot_mid_price:
        mid_prices = order_book.loc[start_index:end_index, ('BIDp1', 'ASKp1')].mean(1)
        ax.step(times, mid_prices.loc[start_index:end_index], 'k-', label='Mid Price')

    ax.set_xlabel('Timestamps')
    ax.set_ylabel('Prices')
    ax.set_title('Evolution Prices ' + ticker)
    ax.legend()
```

#### Plot Volume Evolution over the day

```python
def plot_volume_evolution(msg_book,ticker,ax,mode='executed'):

    opacity = 0.4
    
    # 2 modes to determine 
    if mode == 'executed':
        #  Filter message book on executed orders
        bid_exec = msg_book[(msg_book['Type']==4) & (msg_book['TradeDirection']==1)][['Time','Size']]
        ask_exec = msg_book[(msg_book['Type']==4) & (msg_book['TradeDirection']==-1)][['Time','Size']]
        #  Plot buy sell order
        ax.hist(ask_exec['Time'],bins=40,weights = ask_exec['Size'],color='b', alpha = opacity)
        ax.hist(bid_exec['Time'],bins=40,weights = bid_exec['Size'],color='r', alpha = opacity)
        #  Volumes Traded
        ax.set_title('Trading Volumes ' + ticker)
        ax.legend(labels=['Buy','Sell'],loc=0)
    
    elif mode == 'cancelation':
        #  Filter message book on cancelled orders
        bid_canc = msg_book[(msg_book['Type']>=2) & (msg_book['TradeDirection']==-1) & (msg_book['Type']<=3)][['Time','Size']]
        bid_add = msg_book[(msg_book['Type']==1) & (msg_book['TradeDirection']==-1)][['Time','Size']]
        # Plot
        ax.hist(bid_canc['Time'],bins=40,weights = bid_canc['Size'],color='r', alpha = opacity)
        ax.hist(bid_add['Time'],bins=40,weights = bid_add['Size'],color='b', alpha = opacity)
        #  Volumes Cancelled
        ax.set_title('Canceled Volumes ' + ticker)
        ax.legend(labels=['Bid Cancel','Bid Addition'],loc=0)
        
    
    ax.set_xlabel('Timestamps')
    ax.set_ylabel('Volume')
```

#### Call plotting functions

```python
#  Clear previous plots
plt.clf()
plt.cla()
plt.close()

stocks_ticker = ['AAPL','AMZN','GOOG',"INTC"]

matplotlib.rcParams['figure.figsize'] = [14,12]
fig, ax = plt.subplots(2, 2)
ax = ax.ravel()
plt.subplots_adjust(hspace=.3, wspace =.3)
#plt.subplots_adjust(wspace= 0.5 )

for i in range(len(stocks_ticker)):
    
    ticker = stocks_ticker[i]
    #  Call CSV read
    order_book, msg_book = CSV_read_clean(ticker)
    #  Call plotting functions
    plot_volume_evolution(msg_book,ticker,ax=ax[i],mode='executed')
    #plot_levels_evolution(order_book, msg_book,ax=ax[i],ticker = ticker,nb_levels=1, start_index=1, end_index=500000,plot_mid_price=True)
```
