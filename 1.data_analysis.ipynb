# Data Analysis

#### Import packages 

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
plt.style.use('seaborn-poster')
# print(plt.style.available)
```

#### Define file metadata 

```python
ticker = "AMZN"
sample_date = "2012-06-21"
start_time = 34200000
end_time = 57600000
nb_levels = 10
```

#### Build file names 

```python
order_book_file_name = "_".join([ticker, sample_date, 
                                str(start_time), str(end_time),
                                "orderbook", str(nb_levels)]) + ".csv"
message_book_file_name = "_".join([ticker, sample_date, 
                                  str(start_time), str(end_time), 
                                  "message", str(nb_levels)]) + ".csv"
```

#### Load message order book (data description in the readme) 

```python
msg_book = pd.read_csv(message_book_file_name)
msg_book.columns = ["Time", "Type", 
                    "OrderID", "Size", 
                    "Price", "TradeDirection"]

# msg_book.head()
# msg_book.describe()
# msg_book.shape
```

#### Load order book (data description in the readme)

```python
order_book = pd.read_csv(order_book_file_name)
column_names = []
for i in range(1, nb_levels+1, 1):
    column_names.append("ASKp" + str(i))
    column_names.append("ASKs" + str(i))
    column_names.append("BIDp" + str(i))
    column_names.append("BIDs" + str(i))
order_book.columns = column_names

order_book[list(range(33,40))].head()
# order_book.describe()
# order_book.shape
```

```python
order_book
```

#### Clean data: keep orders emitted during market hours

```python
trading_start_time = 9.5*60*60;       # 9:30:00.000 in ms after midnight
trading_end_time = 16*60*60;        # 16:00:00.000 in ms after midnight

msg_book = msg_book.loc[(msg_book['Time'] >= trading_start_time) &
                        (msg_book['Time'] <= trading_end_time)]

order_book = order_book.loc[(msg_book['Time'] >= trading_start_time) &
                            (msg_book['Time'] <= trading_end_time)]
```

#### Clean data: convert prices to USD 

```python
price_columns = [i for i in order_book.columns.values if 'p' in i]
order_book[price_columns] = order_book[price_columns].apply(lambda x: x/10000, axis=0)
```
